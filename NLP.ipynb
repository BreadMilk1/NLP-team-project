```python
import os
import re
import pandas as pd
import numpy as np
import torch
import torch.nn as nn
import torch.optim as optim
from torch.utils.data import Dataset, DataLoader, Sampler
from sklearn.model_selection import train_test_split
from sklearn.metrics import classification_report, mean_absolute_error
from sklearn.metrics import accuracy_score
import torch.nn.functional as F
from tqdm import tqdm
import warnings
from pathlib import Path
import matplotlib.pyplot as plt
from tokenizers import Tokenizer
from tokenizers.models import WordPiece
from tokenizers.trainers import WordPieceTrainer
from tokenizers.pre_tokenizers import Whitespace
from sklearn.metrics import confusion_matrix, ConfusionMatrixDisplay
import seaborn as sns
import torch.nn.functional as F

STAR_TO_INDEX={
    1:0,2:1,3:2,4:3,7:4,8:5,9:6,10:7
}
INDEX_TO_STAR={v: k for k, v in STAR_TO_INDEX.items()}
# --- 1. Configuration Class for Centralized Parameter Management (XW: Parameter Changes) ---
class CONFIG:
    # !!! IMPORTANT: Modify these paths !!!
    IMDB_DATA_DIR = r"E:\新建文件夹\aclImdb_v1\aclImdb\train"
    GLOVE_FILE_PATH = r"E:\新建文件夹\glove.6B\glove.6B.100d.txt"

    # --- Cache and Tokenizer Paths ---
    CACHE_DIR = Path("./.cache_imdb")
    TOKENIZER_FILE = CACHE_DIR / "wordpiece_tokenizer.json"

    TOKENIZER_VOCAB_SIZE = 30000  # Vocabulary size for the subword tokenizer
    MAX_LEN = 512  # Maximum sequence length (truncation)
    NUM_CLASS=len(STAR_TO_INDEX)
    # --- Model and Training Hyperparameters (Migrated from original code) ---
    # PLAN A- modify
    # SEED = 42
    # EMBED_DIM = 100
    # HIDDEN_DIM = 256
    # NUM_LAYERS = 2
    # DROPOUT_RATE = 0.5
    # LEARNING_RATE = 0.001
    # BATCH_SIZE = 64
    # NUM_EPOCHS = 5
    # NUM_CLASS = 10

      # PLAN C - modify
    SEED = 42
    EMBED_DIM = 100        # Keep consistent with GloVe
    HIDDEN_DIM = 192       # Medium size to balance expressive power and prevent overfitting
    NUM_LAYERS = 1         # Single-layer LSTM to speed up training
    DROPOUT_RATE = 0.5     # Increase dropout to prevent overfitting
    LEARNING_RATE = 0.0005 # Decrease learning rate for more stable training
    BATCH_SIZE = 128       # Increase batch size to accelerate training
    NUM_EPOCHS = 12         # Increase number of training epochs 5
    NUM_CLASS = len(STAR_TO_INDEX)
    PATIENCE = 3           # Early stopping patience

    # PLAN E 
    #SEED = 42
    #EMBED_DIM = 100
    #HIDDEN_DIM = 176       # Medium size to balance expressive power and prevent overfitting
    #NUM_LAYERS = 1         # Single-layer LSTM to speed up training
    #DROPOUT_RATE = 0.55    # Prevent overfitting
    #LEARNING_RATE = 0.0002 # Decrease learning rate for more stable training
    #BATCH_SIZE = 56        # Medium batch size
    #NUM_EPOCHS = 18        # Increase number of training epochs
    #PATIENCE = 5           # Increase early stopping patience
    #GRADIENT_CLIP = 0.8    # Gradient clipping
    #WEIGHT_DECAY = 0.002   # Define weight decay for regularization

    # PLAN F 
    #SEED = 42
    #EMBED_DIM = 100
    #HIDDEN_DIM = 384       # Larger hidden size for more capacity
    #NUM_LAYERS = 2         # Deepen the network
    #DROPOUT_RATE = 0.65    # Stronger dropout to combat overfitting
    #LEARNING_RATE = 0.0001 # Lower learning rate for fine-tuning
    #BATCH_SIZE = 32        # Smaller batch size for stability
    #NUM_EPOCHS = 25        # Longer training
    #PATIENCE = 8           # More patience for early stopping
    #GRADIENT_CLIP = 0.5    # Gradient clipping
    #WEIGHT_DECAY = 0.005   # Define weight decay for regularization
    


    # --- Advanced DataLoader Configuration (Prefetching) ---
    NUM_WORKERS = 0  # Set to 0 on Windows for compatibility, can be 4 or higher on Linux/macOS
    PIN_MEMORY = True


# --- Create Cache Directory ---
CONFIG.CACHE_DIR.mkdir(exist_ok=True)


# --- Device Configuration ---
if torch.backends.mps.is_available():
    DEVICE = torch.device('mps')
elif torch.cuda.is_available():
    DEVICE = torch.device('cuda')
else:
    DEVICE = torch.device('cpu')
print(f"Using device: {DEVICE}")


# --- Random Seed Setup ---
def set_seed(seed):
    torch.manual_seed(seed)
    np.random.seed(seed)
    if torch.cuda.is_available():
        torch.cuda.manual_seed(seed)
        torch.cuda.manual_seed_all(seed)
    torch.backends.cudnn.deterministic = True
    torch.backends.cudnn.benchmark = False


# --- 2. Data Loading ---
def load_data_to_dataframe(data_dir):
    data = []
    for sentiment in ['pos', 'neg']:
        path = os.path.join(data_dir, sentiment)
        if not os.path.exists(path):
            print(f"Warning: Directory {path} not found.")
            continue

        for filename in tqdm(os.listdir(path), desc=f"Loading {sentiment} files"):
            if filename.endswith('.txt'):
                try:
                    rating = int(filename.split('_')[1].split('.')[0])
                    with open(os.path.join(path, filename), 'r', encoding='utf-8') as f:
                        text = f.read()
                    data.append({'review': text, 'sentiment': rating})
                except Exception as e:
                    print(f"Skipping file {filename}: {e}")
    return pd.DataFrame(data)


# --- 3. Text Preprocessing and Subword Tokenization ---
def clean_text_for_tokenizer(text):
    """Performs only basic cleaning, leaving complex tasks to the tokenizer."""
    text = re.sub(r'<[^>]+>', ' ', text)
    return text


def get_tokenizer(data_iterator):
    """Trains a new tokenizer if the file doesn't exist; otherwise, loads from file."""
    if not CONFIG.TOKENIZER_FILE.exists():
        print("Tokenizer cache not found, training a new one...")
        tokenizer = Tokenizer(WordPiece(unk_token="[UNK]"))
        tokenizer.pre_tokenizer = Whitespace()

        trainer = WordPieceTrainer(
            vocab_size=CONFIG.TOKENIZER_VOCAB_SIZE,
            special_tokens=["[PAD]", "[UNK]"]
        )

        tokenizer.train_from_iterator(data_iterator, trainer=trainer)
        tokenizer.save(str(CONFIG.TOKENIZER_FILE))
        print(f"Tokenizer trained and saved to: {CONFIG.TOKENIZER_FILE}")
    else:
        print(f"Loading tokenizer from cache: {CONFIG.TOKENIZER_FILE}")
        tokenizer = Tokenizer.from_file(str(CONFIG.TOKENIZER_FILE))
    return tokenizer


# --- 4. Preprocessing and Caching ---
def preprocess_and_cache(df, tokenizer, name):
    cache_path = CONFIG.CACHE_DIR / f"{name}_data.pt"

    if cache_path.exists():
        print(f"Loading preprocessed '{name}' data from cache: {cache_path}")
        return torch.load(cache_path)

    print(f"Processing and caching '{name}' data...")

    # Batch encoding for efficiency
    cleaned_texts = [clean_text_for_tokenizer(text) for text in df['review'].tolist()]
    encoded_outputs = tokenizer.encode_batch(cleaned_texts)

    processed_data = []
    for i in tqdm(range(len(df)), desc=f"Processing {name} data"):
        ids = encoded_outputs[i].ids

        # Truncation and Padding (to a uniform length)
        if len(ids) > CONFIG.MAX_LEN:
            ids = ids[:CONFIG.MAX_LEN]
        else:
            ids = ids + [tokenizer.token_to_id("[PAD]")] * (CONFIG.MAX_LEN - len(ids))

        label_star = df['sentiment'].iloc[i]   # Labels 1-10 -> 0-9
        label_index=STAR_TO_INDEX.get(label_star)

        processed_data.append({
            'ids': torch.tensor(ids, dtype=torch.long),
            'label': torch.tensor(label_index, dtype=torch.long),
            'length': sum(encoded_outputs[i].attention_mask)  # Save original length for bucketing
        })

    torch.save(processed_data, cache_path)
    print(f"'{name}' data cached to: {cache_path}")
    return processed_data


# --- 5. PyTorch Dataset and DataLoader ---
class PreprocessedIMDbDataset(Dataset):
    """A Dataset that directly uses preprocessed and cached data."""

    def __init__(self, processed_data):
        self.data = processed_data

    def __len__(self):
        return len(self.data)

    def __getitem__(self, idx):
        return self.data[idx]['ids'], self.data[idx]['label']


# --- Bucket Sampler ---
class BucketSampler(Sampler):
    def __init__(self, data_source, batch_size):
        super().__init__()
        self.data_source = data_source
        self.batch_size = batch_size

        # Sort indices based on original length
        lengths = [item['length'] for item in data_source]
        self.sorted_indices = np.argsort(lengths)

    def __iter__(self):
        # Create roughly sorted batches
        batches = [self.sorted_indices[i:i + self.batch_size]
                   for i in range(0, len(self.sorted_indices), self.batch_size)]
        # Shuffle the order of the batches to add randomness to training
        np.random.shuffle(batches)
        # Flatten the list of batches to generate the final index iterator
        for batch in batches:
            yield from batch

    def __len__(self):
        return len(self.data_source)


# --- 6. Helper Function: Load GloVe Vectors ---
def load_pretrained_embeddings(tokenizer, embed_dim, glove_path):
    """
    Accepts a 'tokenizer' object instead of the old 'vocab' object.
    """
    print(f"Loading GloVe embeddings from {glove_path}...")
    try:
        with open(glove_path, 'r', encoding='utf-8') as f:
            glove_vectors = {}
            for line in tqdm(f, desc="Loading GloVe file"):
                parts = line.split()
                word = parts[0]
                # Ensure vector dimension matches
                if len(parts[1:]) == embed_dim:
                    vector = torch.tensor([float(val) for val in parts[1:]])
                    glove_vectors[word] = vector

        print(f"Loaded {len(glove_vectors)} GloVe word vectors.")

        # Use the tokenizer's API
        vocab_size = tokenizer.get_vocab_size()
        weights_matrix = torch.zeros(vocab_size, embed_dim)
        found_words = 0

        # Get vocabulary and indices from the tokenizer
        for word, index in tqdm(tokenizer.get_vocab().items(), desc="Matching vocabulary"):
            if word in glove_vectors:
                weights_matrix[index] = glove_vectors[word]
                found_words += 1
            # Handle subwords, e.g., "playing" can use the vector for "play"
            elif '##' not in word:  # Simple heuristic
                stem = word.rstrip('s').rstrip('ing').rstrip('ed')
                if stem in glove_vectors:
                    weights_matrix[index] = glove_vectors[stem]
                # found_words += 1 # could also be counted as found

        print(f"Found {found_words}/{vocab_size} words in our vocabulary.")
        return weights_matrix

    except FileNotFoundError:
        print(f"Error: GloVe file not found at {glove_path}")
        return None


# --- 7. Deep Learning Model (LSTM + Attention) ---

# --- [ADVANCED UPGRADE: ATTENTION] ---
class Attention(nn.Module):
    """
    Attention mechanism module.
    It receives all time-step outputs from the LSTM and computes a weighted "context vector".
    """

    def __init__(self, hidden_dim):  # hidden_dim will be LSTM's hidden_dim * 2 (because it's bidirectional)
        super(Attention, self).__init__()
        # Weight W
        self.W = nn.Linear(hidden_dim, hidden_dim, bias=False)
        # Context vector v
        self.v = nn.Linear(hidden_dim, 1, bias=False)

    def forward(self, lstm_output, mask):
        # lstm_output shape: [batch_size, seq_len, hidden_dim]
        # mask shape: [batch_size, seq_len, 1]

        # 1. Calculate Energy Scores
        # [B, S, H] -> [B, S, H]
        x = torch.tanh(self.W(lstm_output))
        # [B, S, H] -> [B, S, 1]
        scores = self.v(x)

        # 2. Apply Masking
        # Before softmax, set the scores of all padding positions to a very small value (-1e9)
        # so their weights will be close to 0 after softmax.
        scores = scores.masked_fill(mask == 0, -1e9)

        # 3. Calculate Attention Weights (Alphas)
        # Apply softmax along the seq_len dimension (dim=1)
        # [B, S, 1]
        alphas = F.softmax(scores, dim=1)

        # 4. Calculate Context Vector
        # [B, S, 1] * [B, S, H] -> [B, S, H] (element-wise multiplication)
        context = alphas * lstm_output
        # [B, S, H] -> [B, H] (sum along the seq_len dimension)
        context = torch.sum(context, dim=1)

        return context  # , alphas (can also return alphas if visualization is needed)


# --- [END ADVANCED UPGRADE] ---


class RNNModel(nn.Module):
    def __init__(self, vocab_size, embed_dim, hidden_dim, num_layers,
                 dropout_rate, num_class, pad_idx, embedding_weights=None,
                 freeze_embeddings=False):
        super(RNNModel, self).__init__()
        self.pad_idx = pad_idx  # Save pad_idx for masking

        # --- Embedding Layer ---
        if embedding_weights is None:
            print("Model: Using randomly initialized embedding layer.")
            self.embedding = nn.Embedding(vocab_size, embed_dim, padding_idx=pad_idx)
        else:
            print(f"Model: Using pre-trained embedding layer. Freeze = {freeze_embeddings}")
            self.embedding = nn.Embedding.from_pretrained(
                embedding_weights,
                freeze=freeze_embeddings,
                padding_idx=pad_idx
            )

        # --- LSTM Layer ---
        self.lstm = nn.LSTM(
            input_size=embed_dim,
            hidden_size=hidden_dim,
            num_layers=num_layers,
            batch_first=True,
            dropout=dropout_rate if num_layers > 1 else 0,
            bidirectional=True  # Use bidirectional
        )

        # --- [ADVANCED UPGRADE: ATTENTION] ---
        # Instantiate the attention layer
        # Input dimension is hidden_dim * 2 (because LSTM is bidirectional)
        self.attention = Attention(hidden_dim * 2)
        # --- [END ADVANCED UPGRADE] ---

        self.dropout = nn.Dropout(dropout_rate)

        # --- [MOD] ---
        # Fully-connected classifier
        # Input dimension is still hidden_dim * 2 (from the attention layer's context vector)
        # Output dimension is num_class (now 10)
        self.fc = nn.Linear(hidden_dim * 2, num_class)
        # --- [END MOD] ---

    def forward(self, text):
        # text shape: [batch_size, seq_len]

        # --- [ADVANCED UPGRADE: ATTENTION] ---
        # 1. Create Mask
        # Mask shape is [B, S, 1], used for attention calculation
        # (text != self.pad_idx) will be False (0) at padding positions and True (1) at non-padding positions
        mask = (text != self.pad_idx).unsqueeze(-1)
        # --- [END ADVANCED UPGRADE] ---

        # 2. Embedding
        embedded = self.embedding(text)
        # embedded shape: [batch_size, seq_len, embed_dim]

        # 3. Pass through LSTM
        lstm_out, (hidden, cell) = self.lstm(embedded)
        # lstm_out shape: [batch_size, seq_len, hidden_dim * 2]

        # --- [ADVANCED UPGRADE: ATTENTION] ---
        # 4. Pass through Attention Layer
        # We no longer use the last hidden state (hidden[-1])
        # Instead, we pass all LSTM outputs and the mask to the attention layer
        context_vector = self.attention(lstm_out, mask)
        # context_vector shape: [batch_size, hidden_dim * 2]
        # --- [END ADVANCED UPGRADE] ---

        # 5. Pass through Fully-Connected Layer
        dropped_out = self.dropout(context_vector)  # Apply dropout to the attention output
        return self.fc(dropped_out)  # Output logits


# --- 8. Training and Evaluation Functions (Partially Modified) ---
def train_epoch(model, loader, optimizer, criterion, device):
    model.train()
    total_loss = 0
    for texts, labels in tqdm(loader, desc="Training"):
        texts, labels = texts.to(device), labels.to(device)

        optimizer.zero_grad()
        outputs = model(texts)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        total_loss += loss.item()
    return total_loss / len(loader)


# --- [CORE UPGRADE: FINE-GRAINED] ---
# --- [MODIFIED] evaluate function to return more information: XW ---
def evaluate(model, loader, criterion, device):
    """
    Modified evaluation function that returns raw predictions, labels, and probabilities.
    """
    model.eval()
    total_loss = 0
    all_preds, all_labels, all_probs = [], [], []

    with torch.no_grad():
        for texts, labels in tqdm(loader, desc="Evaluating"):
            texts, labels = texts.to(device), labels.to(device)

            outputs = model(texts) # Logits
            loss = criterion(outputs, labels)
            total_loss += loss.item()

            # Calculate probabilities and predictions
            probs = F.softmax(outputs, dim=1)
            confidence, predicted = torch.max(probs, 1)

            all_preds.extend(predicted.cpu().numpy())
            all_labels.extend(labels.cpu().numpy())
            all_probs.extend(confidence.cpu().numpy()) # Save the max probability for each prediction

    avg_loss = total_loss / len(loader)
    mae = mean_absolute_error(all_labels, all_preds)

    # Return all necessary information
    return avg_loss, mae, all_labels, all_preds, np.array(all_probs)

# [NEW] Evaluation Helper Functions Area
# ==============================================================================

def plot_confusion_matrix(labels, preds, class_names):
    """
    Plots and displays the confusion matrix.
    Corresponds to requirements: accuracy, precision, recall, F1; fine-grained ratings
    """
    cm = confusion_matrix(labels, preds, labels=range(len(class_names)))
    disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=class_names)

    fig, ax = plt.subplots(figsize=(10, 10))
    disp.plot(ax=ax, cmap='Blues', xticks_rotation='vertical')
    ax.set_title("Confusion Matrix")
    plt.tight_layout()
    plt.show()

def analyze_error_buckets(df_test, labels, preds, reviews):
    """
    Analyzes error buckets and prints typical error samples.
    Corresponds to requirement: error buckets
    """
    error_df = pd.DataFrame({
        'review': reviews,
        'true_label': np.array(labels) + 1,  # Convert to 1-10
        'pred_label': np.array(preds) + 1,  # Convert to 1-10
        'abs_error': np.abs(np.array(labels) - np.array(preds))
    })

    # Filter out correctly predicted samples
    error_df = error_df[error_df['abs_error'] > 0].sort_values(by='abs_error', ascending=False)

    print("\n" + "="*20 + " Error Bucket Analysis " + "="*20)

    # 1. Samples with the largest errors
    print("\n--- 1. Large Errors (Prediction-Truth difference > 5) ---")
    large_errors = error_df[error_df['abs_error'] > 5]
    if not large_errors.empty:
        for i, row in large_errors.head(3).iterrows():
            print(f"\nTrue Rating: {row['true_label']}, Predicted Rating: {row['pred_label']}")
            print(f"Review Excerpt: {row['review'][:400]}...")
    else:
        print("Model made no large errors, excellent performance!")

    # 2. Off-by-one errors
    print("\n--- 2. Off-by-One Errors (Prediction-Truth difference is 1) ---")
    off_by_one_errors = error_df[error_df['abs_error'] == 1]
    if not off_by_one_errors.empty:
        for i, row in off_by_one_errors.sample(min(3, len(off_by_one_errors))).iterrows():
            print(f"\nTrue Rating: {row['true_label']}, Predicted Rating: {row['pred_label']}")
            print(f"Review Excerpt: {row['review'][:400]}...")
    else:
        print("Model made no off-by-one errors.")
    print("="*60)

def calculate_ece_and_plot(probs, labels, n_bins=10):
    """
    Calculates Expected Calibration Error (ECE) and plots a reliability diagram.
    Corresponds to requirement: calibration/thresholds
    """
    bin_boundaries = np.linspace(0, 1, n_bins + 1)
    bin_lowers = bin_boundaries[:-1]
    bin_uppers = bin_boundaries[1:]

    confidences, accuracies = [], []
    ece = 0.0

    for bin_lower, bin_upper in zip(bin_lowers, bin_uppers):
        in_bin = (probs > bin_lower) & (probs <= bin_upper)
        prop_in_bin = in_bin.mean()

        if prop_in_bin > 0:
            accuracy_in_bin = labels[in_bin].mean()
            avg_confidence_in_bin = probs[in_bin].mean()

            ece += np.abs(avg_confidence_in_bin - accuracy_in_bin) * prop_in_bin

            confidences.append(avg_confidence_in_bin)
            accuracies.append(accuracy_in_bin)
        else:
            confidences.append(None)
            accuracies.append(None)

    # Plot reliability diagram
    plt.figure(figsize=(8, 8))
    plt.plot([0, 1], [0, 1], linestyle='--', color='gray', label='Perfect Calibration')

    # Filter out empty bins
    valid_indices = [i for i, v in enumerate(confidences) if v is not None]
    valid_confidences = [confidences[i] for i in valid_indices]
    valid_accuracies = [accuracies[i] for i in valid_indices]

    plt.plot(valid_confidences, valid_accuracies, marker='o', linestyle='-', label='Model Calibration')

    plt.xlabel("Confidence (Mean Predicted Probability in Bin)")
    plt.ylabel("Accuracy (Fraction of Positives in Bin)")
    plt.title(f"Reliability Diagram (ECE = {ece:.4f})")
    plt.legend()
    plt.grid(True)
    plt.show()

    return ece

def evaluate_robustness(model, tokenizer, device):
    """
    Evaluates the model on a specially constructed robustness test set.
    Corresponds to requirement: robustness (e.g., negation/sarcasm)
    """
    # This test set needs to be created manually or obtained from other sources. Here we create a few simple examples.
    robustness_data = {
        'review': [
            # Negation
            "This movie was not good, not funny, and not entertaining.", # Expect low score
            "I would not say this is the best film ever, but it is far from the worst.", # Expect neutral or slightly high score
            "Never have I seen such a terrible waste of time.", # Expect low score
            # Sarcasm
            "Wow, another masterpiece of cinema. I was on the edge of my seat... sleeping.", # Expect low score
            "Absolutely brilliant. The director's decision to use a shaky cam for 2 hours was pure genius. My eyes loved it.", # Expect low score
            "I love how the plot just... ended. No resolution, nothing. So artistic." # Expect low score
        ],
        'expected_rating_category': ['low', 'mid', 'low', 'low', 'low', 'low'] # 'low': 1-4, 'mid': 5-7, 'high': 8-10
    }
    df_robust = pd.DataFrame(robustness_data)

    print("\n" + "="*20 + " Robustness Analysis " + "="*20)
    model.eval()
    with torch.no_grad():
        for i, row in df_robust.iterrows():
            text = row['review']
            cleaned_text = clean_text_for_tokenizer(text)
            encoded = tokenizer.encode(cleaned_text)

            ids = torch.tensor([encoded.ids], dtype=torch.long).to(device)

            outputs = model(ids)
            probs = F.softmax(outputs, dim=1)
            pred_label = torch.argmax(probs, dim=1).item() + 1 # 1-10

            print(f"\nText: {text}")
            print(f"Expected: {row['expected_rating_category']} rating category")
            print(f"Model Prediction: {pred_label} stars")
            if (row['expected_rating_category'] == 'low' and pred_label > 4) or \
               (row['expected_rating_category'] == 'high' and pred_label < 8):
                print("--> Result: Model may have failed to correctly understand negation/sarcasm.")
            else:
                print("--> Result: Model performed as expected.")
    print("="*65)

# ==============================================================================

# --- [END CORE UPGRADE] ---


# --- 9. Main Execution Script ---
if __name__ == "__main__":
    set_seed(CONFIG.SEED)

    # --- Step 1 & 2: Load and Split Data ---
    print("Loading raw data...")
    all_train_df = load_data_to_dataframe(CONFIG.IMDB_DATA_DIR)

    if all_train_df.empty:
        raise SystemExit(f"Error: Could not load data from {CONFIG.IMDB_DATA_DIR}. Please check the path.")

    train_val_df, test_df = train_test_split(all_train_df, test_size=0.2, random_state=CONFIG.SEED)
    train_df, val_df = train_test_split(train_val_df, test_size=0.25, random_state=CONFIG.SEED)

    print(f"\nTraining samples: {len(train_df)}")
    print(f"Validation samples: {len(val_df)}")
    print(f"Test samples: {len(test_df)}")

    # --- Step 3: Train/Load Tokenizer ---
    text_iterator = (clean_text_for_tokenizer(text) for text in train_df['review'])
    tokenizer = get_tokenizer(text_iterator)

    # --- Step 4: Update Global Variables ---
    VOCAB_SIZE = tokenizer.get_vocab_size()
    PAD_IDX = tokenizer.token_to_id("[PAD]")
    print(f"Vocabulary Size: {VOCAB_SIZE}, PAD Index: {PAD_IDX}")

    # --- Step 5: Preprocess and Cache Data ---
    train_processed = preprocess_and_cache(train_df, tokenizer, 'train')
    val_processed = preprocess_and_cache(val_df, tokenizer, 'val')
    test_processed = preprocess_and_cache(test_df, tokenizer, 'test')

    train_dataset = PreprocessedIMDbDataset(train_processed)
    val_dataset = PreprocessedIMDbDataset(val_processed)
    test_dataset = PreprocessedIMDbDataset(test_processed)

    # --- Step 6: Create DataLoaders with Bucketing and Prefetching ---
    train_sampler = BucketSampler(train_processed, batch_size=CONFIG.BATCH_SIZE)

    # Note: shuffle must be False because BucketSampler already handles randomization
    train_loader = DataLoader(
        train_dataset,
        batch_size=CONFIG.BATCH_SIZE,
        sampler=train_sampler,
        num_workers=CONFIG.NUM_WORKERS,
        pin_memory=CONFIG.PIN_MEMORY
    )
    val_loader = DataLoader(
        val_dataset,
        batch_size=CONFIG.BATCH_SIZE,
        shuffle=False,
        num_workers=CONFIG.NUM_WORKERS,
        pin_memory=CONFIG.PIN_MEMORY
    )
    test_loader = DataLoader(
        test_dataset,
        batch_size=CONFIG.BATCH_SIZE,
        shuffle=False,
        num_workers=CONFIG.NUM_WORKERS,
        pin_memory=CONFIG.PIN_MEMORY
    )

    # --- Step 7: Load GloVe ---
    pretrained_weights = load_pretrained_embeddings(tokenizer, CONFIG.EMBED_DIM, CONFIG.GLOVE_FILE_PATH)

    # --- Step 8: Experiment A (Train from Scratch) ---
    print("\n" + "=" * 30)
    print(" Starting Experiment A: Train Embedding from Scratch (LSTM+Attention)")
    print("=" * 30)
    model_A = RNNModel(
        vocab_size=VOCAB_SIZE,
        embed_dim=CONFIG.EMBED_DIM,
        hidden_dim=CONFIG.HIDDEN_DIM,
        num_layers=CONFIG.NUM_LAYERS,
        dropout_rate=CONFIG.DROPOUT_RATE,
        num_class=CONFIG.NUM_CLASS,  # 10 classes
        pad_idx=PAD_IDX,
        embedding_weights=None
    ).to(DEVICE)

    optimizer_A = optim.Adam(model_A.parameters(), lr=CONFIG.LEARNING_RATE)
    criterion = nn.CrossEntropyLoss()  # CrossEntropyLoss is perfect for multi-class classification

    for epoch in range(CONFIG.NUM_EPOCHS):
        print(f"\n--- Epoch {epoch + 1}/{CONFIG.NUM_EPOCHS} (Experiment A) ---")
        train_loss = train_epoch(model_A, train_loader, optimizer_A, criterion, DEVICE)

        # --- [CORE UPGRADE: FINE-GRAINED] ---
        # [FIX] Add an underscore _ to receive and ignore val_probs: XW
        val_loss, val_mae, val_labels, val_preds, _ = evaluate(model_A, val_loader, criterion, DEVICE)
        print(f"Epoch {epoch + 1} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val MAE: {val_mae:.4f} (stars)")
        val_accuracy = accuracy_score(val_labels, val_preds)
        print(f" Val Accuracy (not recommended, for reference only): {val_accuracy:.4f}")
        # --- [END CORE UPGRADE] ---

    # --- Step 9: Experiment B (Pre-trained + Fine-tuning) ---
    model_B = None
    if pretrained_weights is not None:
        print("\n" + "=" * 30)
        print(" Starting Experiment B: Pre-trained Embedding (LSTM+Attention)")
        print("=" * 30)

        model_B = RNNModel(
            vocab_size=VOCAB_SIZE,
            embed_dim=CONFIG.EMBED_DIM,
            hidden_dim=CONFIG.HIDDEN_DIM,
            num_layers=CONFIG.NUM_LAYERS,
            dropout_rate=CONFIG.DROPOUT_RATE,
            num_class=CONFIG.NUM_CLASS,  # 10 classes
            pad_idx=PAD_IDX,
            embedding_weights=pretrained_weights.to(DEVICE),
            freeze_embeddings=False
        ).to(DEVICE)

        optimizer_B = optim.Adam(model_B.parameters(), lr=CONFIG.LEARNING_RATE)
        # criterion remains the same

        for epoch in range(CONFIG.NUM_EPOCHS):
            print(f"\n--- Epoch {epoch + 1}/{CONFIG.NUM_EPOCHS} (Experiment B) ---")
            train_loss = train_epoch(model_B, train_loader, optimizer_B, criterion, DEVICE)

            # --- [CORE UPGRADE: FINE-GRAINED] ---
            # [FIX] Add an underscore _ to receive and ignore val_probs
            val_loss, val_mae, val_labels, val_preds, _ = evaluate(model_B, val_loader, criterion, DEVICE)
            print(
                f"Epoch {epoch + 1} | Train Loss: {train_loss:.4f} | Val Loss: {val_loss:.4f} | Val MAE: {val_mae:.4f} (stars)")
            val_accuracy = accuracy_score(val_labels, val_preds)  # [FIX] Manually calculate accuracy
            print(f" Val Accuracy (not recommended, for reference only): {val_accuracy:.4f}")
            # --- [END CORE UPGRADE] ---
    else:
        print("\n" + "=" * 30)
        print("Skipping Experiment B because GloVe weights were not loaded.")

    # --- Step 10: Final Evaluation (on Test Set) ---
print("\n" + "=" * 40)
print(" Final Evaluation (on Test Set)")
print("=" * 40)

# Prepare original review texts for error analysis
test_reviews = test_df['review'].tolist()
# Define rating names
class_names = [str(INDEX_TO_STAR[i]) for i in range(CONFIG.NUM_CLASS)]

# --- [Systematic Evaluation] ---
# Package models and their names for easy looping, a simple form of systematic search/ablation study
models_to_evaluate = {
    "Model A (Train from Scratch)": model_A,
    # "Model B (Pre-trained + Fine-tune)": model_B # Add model_B to evaluation if it exists
}
if model_B is not None:
    models_to_evaluate["Model B (Pre-trained + Fine-tune)"] = model_B

# Create a DataFrame to record experiment results, part of the Systematic Search
results_summary = []

for model_name, model in models_to_evaluate.items():
    print(f"\n\n{'='*25} Evaluating: {model_name} {'='*25}")

    # 1. Get basic metrics (MAE, Loss) and detailed prediction results
    #    Corresponds to requirement: fine-grained ratings
    # Code for the final evaluation part (this is correct)
    test_loss, test_mae, test_labels, test_preds, test_probs = evaluate(model, test_loader, criterion, DEVICE)

    print(f"\n{model_name} | Test Loss: {test_loss:.4f}")
    print(f"{model_name} | ★★★ Test MAE (Mean Absolute Error): {test_mae:.4f} (stars) ★★★")

    # 2. Print detailed classification report (Precision, Recall, F1)
    #    Corresponds to requirements: accuracy, precision, recall, F1
    unique_labels_in_data = sorted(list(set(test_labels)))
    
    target_names_for_report = [str(INDEX_TO_STAR[label]) for label in unique_labels_in_data]
    test_report_str = classification_report(test_labels, test_preds, labels=unique_labels_in_data, target_names=target_names_for_report, zero_division=0)
    print(f"\n{model_name} | Detailed Classification Report (Labels 1,2,3,4,7,8,9,10):")
    print(test_report_str)

    # 3. Plot confusion matrix
    #    Corresponds to requirements: accuracy, ..., F1; fine-grained ratings
    print(f"\n{model_name} | Generating Confusion Matrix...")
    plot_confusion_matrix(test_labels, test_preds, class_names)

    # 4. Analyze error buckets
    #    Corresponds to requirement: error buckets
    analyze_error_buckets(test_df, test_labels, test_preds, test_reviews)

    # 5. Calibration Analysis
    #    Corresponds to requirement: calibration/thresholds
    print(f"\n{model_name} | Performing Calibration Analysis...")
    # To calculate calibration, we need a binary label: was the prediction correct?
    is_correct = (np.array(test_preds) == np.array(test_labels)).astype(int)
    ece = calculate_ece_and_plot(test_probs, is_correct)
    print(f"{model_name} | Expected Calibration Error (ECE): {ece:.4f} (lower is better)")

    # 6. Robustness Analysis
    #    Corresponds to requirement: robustness (e.g., negation/sarcasm)
    evaluate_robustness(model, tokenizer, DEVICE)

    # 7. Record experiment results (Systematic Search)
    #    Corresponds to requirement: Systematic search/ablations
    accuracy = accuracy_score(test_labels, test_preds)
    results_summary.append({
        "Model": model_name,
        "MAE": test_mae,
        "Accuracy": accuracy,
        "ECE": ece,
        "Learning Rate": CONFIG.LEARNING_RATE, # Record key hyperparameters
        "Dropout": CONFIG.DROPOUT_RATE,
        "Hidden Dim": CONFIG.HIDDEN_DIM
    })

# --- Step 11: Print Final Systematic Comparison Report ---
print("\n\n" + "=" * 30)
print(" Systematic Experiment Results Summary ")
print("=" * 30)
summary_df = pd.DataFrame(results_summary)
print(summary_df.to_string())
        # --- [END CORE UPGRADE] ---







```
